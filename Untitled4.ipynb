```python
import tensorflow as tf
import numpy as np
import pandas as pd

data = pd.read_csv('blood.csv')

# Sigmoid function using TensorFlow
def sigmoid(x):
    return 1 / (1 + tf.exp(-x))  # Use TensorFlow's exp function

# Logistic Regression class in TensorFlow (Using tf.keras.Model)
class My_LogisticRegression_Model(tf.keras.Model):
    def __init__(self, lr, n_iters,reg_lambda=0.01):
        super(My_LogisticRegression_Model, self).__init__()
        self.lr = lr  # Learning rate
        self.n_iters = n_iters  # Number of iterations for training
        self.reg_lambda = reg_lambda  # Regularization strength
        self.optimizer = tf.optimizers.Adam(learning_rate=self.lr)  # Optimizer

    def build(self, input_shape):
        n_features = input_shape[1]
       # Initialize weights and bias using add_weight (don't manually assign to self.weights)
        self.add_weight(name="weights", shape=(n_features,), initializer="zeros", trainable=True)
        self.add_weight(name="bias", shape=(), initializer="zeros", trainable=True)

    def call(self, inputs):
 # Access weights and bias from the model's trainable variables
        weights, bias = self.trainable_variables
        
        # Linear prediction: weights * inputs + bias
        linear_pred = tf.reduce_sum(tf.multiply(inputs, weights), axis=1) + bias
        predictions = sigmoid(linear_pred)
        return predictions

    def fit(self, X, y):
        # Convert data to TensorFlow tensors
        X = tf.convert_to_tensor(X, dtype=tf.float32)
        y = tf.convert_to_tensor(y, dtype=tf.float32)

        # Training loop
        for _ in range(self.n_iters):
            with tf.GradientTape() as tape:
                predictions = self(X)
                # Loss function (binary cross-entropy)
                loss = tf.reduce_mean(tf.nn.sigmoid_cross_entropy_with_logits(labels=y, logits=predictions))

            # Compute gradients and apply them
            gradients = tape.gradient(loss, self.trainable_variables)
            self.optimizer.apply_gradients(zip(gradients, self.trainable_variables))

    def predict(self, X):
        X = tf.convert_to_tensor(X, dtype=tf.float32)
        predictions = self(X)
        class_pred = [0 if y <= 0.5 else 1 for y in predictions.numpy()]  # Convert to class labels (0 or 1)
        return class_pred






from sklearn.model_selection import train_test_split
# Features and target
X = df[['Months since Last Donation', 'Number of Donations', 'Total Volume Donated (c.c.)', 'Months since First Donation']].values
y = df['Made Donation this month'].values

# Split into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1234)



from sklearn.preprocessing import StandardScaler

# Standardizing the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)

X_test_scaled = scaler.transform(X_test)



# Initialize and train the logistic regression model
model = My_LogisticRegression_Model(lr=0.001,n_iters=2000)
model.compile(optimizer=model.optimizer, loss=tf.losses.BinaryCrossentropy(from_logits=True))
model.fit(X_train_scaled, y_train)

# Make predictions
y_pred = model.predict(X_test_scaled)

# Calculate accuracy
def accuracy(y_pred, y_test):
    return np.sum(y_pred == y_test) / len(y_test)

# Print accuracy
acc = accuracy(y_pred, y_test)
print(f"Model Accuracy: {acc}")



# Define input signature
@tf.function(input_signature=[tf.TensorSpec(shape=[None, X_train.shape[1]], dtype=tf.float32)])
def predict_fn(inputs):
    return model.predict(inputs)

# Save the model as a TensorFlow SavedModel format
tf.saved_model.save(model, 'saved_model/logistic_regression_model')








import tensorflow as tf
# Convert the SavedModel to TensorFlow Lite format
converter = tf.lite.TFLiteConverter.from_saved_model('/content/saved_model/logistic_regression_model')
tflite_model = converter.convert()


# Save the TFLite model to a file
tflite_model_path = 'logistic_regression_model.tflite'
with open(tflite_model_path, 'wb') as f:
    f.write(tflite_model)

print("Model saved and converted to TFLite format.")

```
